```python
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(12, 15))


x = np.array([80, 85, 90, ])
y = np.array([240, 250, 260, ])

# Subplot 1
plt.subplot(3, 3, 1)
plt.plot(x, y, 'o')
plt.title("Title 1")
plt.xlabel("x label")
plt.ylabel("y label")
#plt2


x = np.array([2, 6, 3, 5])
y = np.array([5, 2, 5, 2])

plt.subplot(3, 3, 2)

plt.plot(x,c='orange',marker='*')
plt.plot(y,ls =':',c='blue')


#plt3


x = np.array([2, 4, 3, 5])
plt.subplot(3, 3, 3)
plt.plot(x, marker = 'o', ms = 20, mfc = 'y',mec='black')
#plt4


x = np.array([10,20,30,40,50])
y = np.array([110,120,130,140,150])

plt.title("Title")

plt.subplot(3, 3, 4)
plt.plot(x,y)


plt.grid()


#plt 5
plt.subplot(3, 3, 5)
x = np.array([5,7,8,7,2])
y = np.array([99,86,87,88,111])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15])
y = np.array([12,34,43,23,56])
sizes = np.array([20,50,100,200,500])

plt.scatter(x, y, s=sizes, color = 'darkgreen',alpha=0.5)
plt.subplot(3, 3, 5)

#plt6
plt.subplot(3, 3, 6)
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y,color='green', width = 0.1)
plt.show()




# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
# Plot pie chart
plt.pie(y, labels=mylabels, explode=myexplode, startangle=90, shadow=True, autopct='%1.1f%%')
# Place legend outside the pie chart
plt.legend(title="Four Fruits:", bbox_to_anchor=(1.05, 1), loc='upper left')

# Overall title
plt.suptitle("GRAPH", fontsize=20)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=0.8, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

# Show plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Increase figure size
plt.figure(figsize=(12, 12))

# Graph 1
x = np.array([2, 4])
y = np.array([3, 6])

plt.subplot(3, 3, 1)
plt.title('Title 1')
plt.xlabel("X label", color="blue")
plt.ylabel("Y label", color="blue")
plt.plot(x, y, 'o', color="blue")

# Graph 2
y1 = np.array([3, 8, 1, 9])
y2 = np.array([6, 2, 8, 3])

plt.subplot(3, 3, 2)
plt.plot(y1, color="orange", marker='*')
plt.plot(y2, linewidth=1.5, color="blue", linestyle="dashed")

# Graph 3
ypoints = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='o', ms=10, mec="black", mfc="yellow")

# Graph 4
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([210, 220, 230, 240, 250, 260, 270, 280, 290, 300])

plt.subplot(3, 3, 4)
plt.title("Title 2")
plt.plot(x, y)
plt.grid()

# Graph 5
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
colors = np.random.randint(50, size=(50))
sizes = 1 * np.random.randint(50, size=(50))

plt.subplot(3, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

# Graph 6
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 6)
plt.bar(x, y, color="red", width=0.1)

# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([35, 25, 25, 15])
mylabels = ["mlp", "pkd", "kkz", "mkd"]
myexplode = [0.2, 0, 0, 0]
# Plot pie chart
plt.pie(y, labels=mylabels, explode=myexplode, startangle=90, shadow=True, autopct='%1.1f%%')
# Place legend outside the pie chart
plt.legend(title="Districts:", bbox_to_anchor=(1.05, 1), loc='upper left')

# Overall title
plt.suptitle("GRAPH", fontsize=20)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=0.8, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

# Show plot
plt.show()
```
