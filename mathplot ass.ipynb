```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(12, 15))


x = np.array([80, 85, 90, ])
y = np.array([240, 250, 260, ])

# Subplot 1
plt.subplot(3, 3, 1)
plt.plot(x, y, 'o')
plt.title("Title 1")
plt.xlabel("x label")
plt.ylabel("y label")
#plt2


x = np.array([2, 6, 3, 5])
y = np.array([5, 2, 5, 2])

plt.subplot(3, 3, 2)

plt.plot(x,c='orange',marker='*')
plt.plot(y,ls =':',c='blue')


#plt3


x = np.array([2, 4, 3, 5])
plt.subplot(3, 3, 3)
plt.plot(x, marker = 'o', ms = 20, mfc = 'y',mec='black')
#plt4


x = np.array([10,20,30,40,50])
y = np.array([110,120,130,140,150])

plt.title("Title")

plt.subplot(3, 3, 4)
plt.plot(x,y)


plt.grid()


#plt 5
plt.subplot(3, 3, 5)
x = np.array([5,7,8,7,2])
y = np.array([99,86,87,88,111])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15])
y = np.array([12,34,43,23,56])
sizes = np.array([20,50,100,200,500])

plt.scatter(x, y, s=sizes, color = 'darkgreen',alpha=0.5)
plt.subplot(3, 3, 5)

#plt6
plt.subplot(3, 3, 6)
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y,color='red', width = 0.1)
plt.show()

#plt7
plt.subplot(3, 3, 7)
y = np.array([35, 25, 20, 25])
myexplode = [0, 0, 0, 0.2]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
# Subplot 7

plt.pie(y, startangle=90, explode=myexplode, shadow=True, colors=mycolors,labels = mylabels)
plt.title('Pie Chart')



# Adjust layout to prevent overlap
plt.tight_layout()

plt.show() 
```
